```python
import pandas as pd
```

```python
df = pd.read_csv('energy.csv')
df[df["Year"]==2019]["Expenditure.Commercial.Coal"]
```

```python
#firsttwo = df[["State","Year"]]
expDf = df.filter(like='Expenditure') 
expDf =expDf.assign(State = df["State"],Year = df["Year"])
first = expDf.pop("State")
two = expDf.pop("Year")
expDf.insert(0,"State",first)
expDf.insert(1,"Year",two)
expDf
```

```python
expDf = expDf.loc[df['Year']==2019]
#print(expDf.filter(like = "Expenditure").sum(1))
expDf['Expenditure.Total'] = expDf.filter(like='Expenditure').sum(1)
expDf.head()
```

```python
expDf = expDf.sort_values(by=['Expenditure.Total'])
expDf
```

```python
expDf = expDf[:10]
expDf
```

```python
df.drop(df.loc[df['Year']!=2019].index, inplace=True)

df['Consumption.Total'] = df.filter(like='Consumption').sum(1)
df
```

```python
df.to_csv('energy_updated.csv', index=False)  
```
